```python
import numpy as np  # For numerical operations
import math as mt  # For mathematical functions
import pandas as pd  # For data manipulation
import random  # For random number generation
import matplotlib.pyplot as plt  # For plotting
from sklearn.metrics import silhouette_score  # For evaluating clustering performance
from sklearn.preprocessing import StandardScaler  # For data normalization
from pprint import pprint  # For pretty-printing data structures
```

```python
import math as mt
from queue import PriorityQueue

# Class definitions
class SDN:
    def __init__(self, index, x, y, CPUcyc):
        self.id = index
        self.x = x
        self.y = y
        self.tasks_incoming = []
        self.tasks_priority_queue = PriorityQueue()
        self.fcfs_queue = PriorityQueue()
        self.sjf_queue = PriorityQueue()
        self.list_of_fns = []
        self.list_of_eus = []
        self.CPUcyc = CPUcyc

class FN:
    def __init__(self, index, x, y, freq, CPUcyc, failrate):
        self.id = index
        self.x = x
        self.y = y
        self.sdn_num = None
        self.freq = freq
        self.CPUcyc = CPUcyc
        self.failrate = failrate

class EU:
    def __init__(self, index, x, y):
        self.id = index
        self.x = x
        self.y = y
        self.sdn_num = None
        self.tasks = []

# Constants
N = 5000  # Number of end users
M = 100   # Number of fog nodes

# Lists
fns = []  # List of all fog nodes
fns_pos = []  # Positions of fog nodes
eus = []  # List of all end users
eus_pos = []  # Positions of end users

# Failure rate calculation for fog nodes using a Poisson distribution
def computeFailRate():
    _lambda = 25
    p = random.random()
    return -(mt.log(1.0 - p) / _lambda)

# Initialize fog nodes with random attributes
for i in range(M):
    x, y = random.random() * 1000000, random.random() * 1000000
    freq, CPUcyc = random.randint(1, 15), random.randint(500, 1000)
    failrate = computeFailRate()
    fns.append(FN(i, x, y, freq, CPUcyc, failrate))
    fns_pos.append([x, y])

# Initialize end users with random positions
for i in range(N):
    x, y = random.random() * 1000000, random.random() * 1000000
    eus.append(EU(i, x, y))
    eus_pos.append([x, y])

# Plotting
plt.figure(figsize=(7, 6), facecolor='white')
fn_np = np.array(fns_pos)
eu_np = np.array(eus_pos)
plt.scatter(fn_np[:, 0], fn_np[:, 1], color='green', marker='o', s=50, label='Fog Nodes')
# plt.scatter(eu_np[:, 0], eu_np[:, 1], color='blue', marker='o', s=10, label='End Users')
plt.title('Unclustered Fog Nodes and End Users')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
import numpy as np

# Settings and variables
max_clusters = int(M / 4)  # Reduce the number of clusters to a quarter of the number of nodes
sse = []  # List to hold the sum of squared errors for each k value

# Perform K-means clustering for a range of cluster counts to find the elbow
for k in range(1, max_clusters):
    kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
    kmeans.fit(fn_np)  # Assuming fn_np is your array of fog node positions
    sse.append(kmeans.inertia_)  # Append the sum of squared errors

# Plotting the SSE to find the elbow
plt.figure(figsize=(7, 6))
plt.plot(range(1, max_clusters), sse, marker='o')
plt.title('Elbow Method for Determining Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.grid(True)
plt.xticks(range(1, max_clusters))
plt.style.use("fivethirtyeight")

# Identifying the elbow point using KneeLocator
kl = KneeLocator(range(1, max_clusters), sse, curve='convex', direction='decreasing')
elbow_point = kl.elbow
print(f"Optimal number of clusters (Elbow Method): {elbow_point}")

# Highlight the elbow point on the plot
if elbow_point is not None:
    plt.axvline(x=elbow_point, color='r', linestyle='--', label=f'Elbow at k={elbow_point}')
    plt.legend()

plt.show()
```

```python
#Generate tasks
import sched, time
from time import mktime
import time
# Class for tasks with proper initialization and less-than operator for priority queuing
class Task:
    def __init__(self, id, gen_eu, gen_sdn, gen_time, deadline, size, freq, CPUcap):
        self.id = id
        self.gen_eu = gen_eu
        self.gen_sdn = gen_sdn
        self.gen_time = gen_time
        self.deadline = deadline
        self.size = size
        self.freq = freq
        self.CPUcap = CPUcap

    def __lt__(self, other):
        return self.id < other.id

# Task generation based on an exponential distribution for arrival times
def generate_tasks(sdns, eus, _lambda=5):
    tasks = []
    arrival_time = time.time()
    count = 0

    for sdn in sdns:
        sdn.tasks_incoming.clear()

    for eu in eus:
        eu.tasks.clear()

    for sdn in sdns:
        for eu in sdn.list_of_eus:
            num_tasks = random.randint(0, 3)  # Each EU generates 0 to 3 tasks
            for _ in range(num_tasks):
                p = random.random()
                inter_arrival_time = -mt.log(1.0 - p) / _lambda
                arrival_time += inter_arrival_time
                task = Task(count, eu.id, sdn.id, arrival_time, arrival_time + random.uniform(250, 500), random.randint(500, 2000), 250, random.randint(500, 1000))
                eu.tasks.append(task)
                sdn.tasks_incoming.append(task)
                tasks.append(task)
                count += 1

    return tasks, count

tasks, total_tasks = generate_tasks(sdns, eus)
print("Count of tasks is", total_tasks)

# Example for printing the number of tasks for each SDN
for sdn in sdns:
    print(f"SDN {sdn.id} has {len(sdn.tasks_incoming)} incoming tasks.")
```

```python
def computeQueueingDelay(task, sdn):
    """Calculate the queuing delay of a task based on its frequency, position in the queue, and the SDN's CPU cycle capacity."""
    if sdn.CPUcyc == 0:
        raise ValueError("SDN CPU cycle capacity cannot be zero.")
    return task.freq * task.queue_pos / sdn.CPUcyc

def computeOffloadingDelay(task, fn):
    """Calculate the offloading delay of a task based on its size, frequency, and the fog node's CPU cycle capacity."""
    if fn.CPUcyc == 0:
        raise ValueError("Fog node CPU cycle capacity cannot be zero.")
    return task.size * task.freq / fn.CPUcyc

def computeTransmissionDelay(task, fn):
    """Compute the transmission delay for a task from an end user to a fog node."""
    gen_eu = eus[task.gen_eu]  # Assuming 'eus' is defined elsewhere and accessible
    ch = Channel()
    return ch.get_bw(0.5, gen_eu.x, gen_eu.y, fn.x, fn.y)
```

```python
class Channel:
    W = math.pow(10,7)  # Bandwidth
    No = 12.59 * math.pow(10,-15)  # Noise power spectral density
    h = 0.50  # Path loss exponent?
    p0 = 0.758  # Path loss intercept
    gamma = 0.01  # Path loss slope
    d0 = 1000  # Reference distance
    interPower = 50.119 * math.pow(10,-9)  # Interference power (-109dBm)
    shadowFactor = 0.3162  # Shadowing factor
    commFailure = 0.002  # Communication failure rate

    def get_pathloss(self, x1, y1, x2, y2):
        """Calculate the path loss based on the log-distance path loss model."""
        d = max(math.sqrt((x2 - x1)**2 + (y2 - y1)**2), self.d0)  # Ensure d is at least d0
        return self.p0 + (10 * self.gamma * math.log10(d / self.d0))

    def get_bw(self, power, x1, y1, x2, y2):
        """Calculate the available bandwidth based on the given power and path loss."""
        pathLoss = self.get_pathloss(x1, y1, x2, y2)
        return math.log2(1 + ((self.shadowFactor * power * pathLoss) / (self.interPower + (self.W * self.No))))
```

```python
sdn = sdns[0]
num_tasks = 1000 # the number of tasks we are taking from the priority queue at any instance and assigning it

connections = {}
for i in range(num_tasks):
      task = sdn.tasks_priority_queue.get()[1]
      allowed_delay = task.deadline-task.gen_time
      task.queue_pos = i+1
      connections[i] = {}
      connections[i]['task_id'] = task.id
      connections[i]['fns'] = []
      qd = computeQueueingDelay(task, sdn)
      for fn in sdn.list_of_fns:
          if(fn.failrate > 0.05):
            pass
          od = computeOffloadingDelay(task, fn)
          td = computeTransmissionDelay(task, fn)
          total_delay = qd + od + td
          # print("qd " + str(qd))
          # print("od " + str(od))
          # print("td " + str(td))
          # print("total " + str(total_delay))
          if(total_delay <= allowed_delay):
            connections[i]['fns'].append(fn.id)

  # for item in connections.items():
  #   print(item)
```

```python
# Minimize function
def fitness_func(args):
    pop = args
    cost = 0
    for task_idx, fn_idx in enumerate(pop):
        if(len(connections[task_idx]['fns'])) != 0:
            task = tasks[connections[task_idx]['task_id']]
            fn_idx = min(int(fn_idx), len(connections[task_idx]['fns'])-1 )
            fn = fns[connections[task_idx]['fns'][int(fn_idx)]];
            delay = computeOffloadingDelay(task, fn) + computeQueueingDelay(task, sdns[0]) + computeTransmissionDelay(task, fn)
            cost += 0.5*delay + 0.5*fn.failrate
    return cost

def find_max_cost(args):
    pop = args
    mx = 0
    for task_idx, fn_idx in enumerate(pop):
        if(len(connections[task_idx]['fns'])) != 0:
            task = tasks[connections[task_idx]['task_id']]
            fn_idx = min(int(fn_idx), len(connections[task_idx]['fns'])-1 )
            fn = fns[connections[task_idx]['fns'][int(fn_idx)]];
            delay = computeOffloadingDelay(task, fn) + computeQueueingDelay(task, sdns[0]) + computeTransmissionDelay(task, fn)
            mx = max(mx, delay)
    return mx
def generate_initial_population(num_individuals, tasks, task_options_by_task):
    initial_population = []
    for _ in range(num_individuals):
        individual = []
        for task_options in task_options_by_task:
            if task_options:  # Ensure there are available options
                fn_idx = random.randint(0, len(task_options) - 1)
                individual.append(fn_idx)
            else:
                individual.append(-1)  # No options available
        initial_population.append(individual)
    return initial_population

# Example usage
task_options_by_task = [connections[j]['fns'] if connections[j]['fns'] else [-1] for j in range(num_tasks)]
common_initial_population = generate_initial_population(50, num_tasks, task_options_by_task)
```

```python
class GeneticAlgorithm:
    def __init__(self, pop_size=1, tasks=1, max_iter=100, mutation_rate=0.1):
        self.cost_func = None
        self.pop_size = pop_size
        self.tasks = tasks
        self.max_iter = max_iter
        self.mutation_rate = mutation_rate
        self.population =  common_initial_population.copy()
        self.best_individual = None
        self.best_fitness = float('inf')
        self.avg_fitness = []
        self.best_fitnesses = []

    # def initialize_population(self):
    #     for i in range(self.pop_size):
    #         particle_position = []
    #         for j in range(self.tasks):
    #             if connections[j]['fns']:  # Check if fog nodes are available for this task
    #                 fn_idx = random.randint(0, len(connections[j]['fns']) - 1)
    #                 particle_position.append(fn_idx)
    #             else:
    #                 particle_position.append(-1)
    #         self.population.append(particle_position)

    def evaluate_population(self):
        fitness_values = np.zeros(self.pop_size)
        for i in range(self.pop_size):
            fitness_values[i] = self.cost_func(self.population[i])
        return fitness_values

    def crossover(self, parent1, parent2):
        crossover_point = np.random.randint(1, self.tasks)
        child = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        return child

    def mutate(self, individual):
        for i in range(self.tasks):
            fn_length = len(connections[i]['fns'])
            if fn_length > 0 and np.random.rand() < self.mutation_rate:
                individual[i] = np.random.randint(0, fn_length)
        return individual


    def evolve(self):
        new_population = np.zeros((self.pop_size, self.tasks))
        fitness_values = self.evaluate_population()
        
        avg_fitness = np.mean(fitness_values)
        best_fitness = np.min(fitness_values)
        if best_fitness < self.best_fitness:
            self.best_fitness = best_fitness
            self.best_individual = self.population[np.argmin(fitness_values)]
        self.avg_fitness.append(avg_fitness)
        self.best_fitnesses.append(self.best_fitness)

        for i in range(self.pop_size):
            parent1 = self.population[np.random.randint(0, self.pop_size)]
            parent2 = self.population[np.random.randint(0, self.pop_size)]
            child = self.crossover(parent1, parent2)
            child = self.mutate(child)
            new_population[i] = child

        self.population = new_population
        # fitness_values = self.evaluate_population()
        # avg_fitness = np.mean(fitness_values)
        # best_fitness = np.min(fitness_values)
        # if best_fitness < self.best_fitness:
        #     self.best_fitness = best_fitness
        #     self.best_individual = self.population[np.argmin(fitness_values)]
        # self.avg_fitness.append(avg_fitness)
        # self.best_fitnesses.append(self.best_fitness)

    def start(self):
        # self.initialize_population()
        for _ in range(self.max_iter):
            self.evolve()
        plt.figure(figsize=(8, 5))
        plt.plot(self.best_fitnesses, color='darkviolet')
        plt.title('Best Fitness over Generations')
        plt.xlabel('Generation')
        plt.ylabel('Best Fitness')
        plt.show()
        # print("Optimal assignment:", self.best_individual)
        # print("Minimum cost:", self.best_fitness)

# Initialize and run GA
ga = GeneticAlgorithm(pop_size=50, tasks=num_tasks, max_iter=70, mutation_rate=0.1)
ga.cost_func = fitness_func
ga.start()
```

```python
class JayaAlgorithm:
    def __init__(self, pop_size=1, tasks=1, max_iter=100):
        self.cost_func = None
        self.pop_size = pop_size
        self.tasks = tasks
        self.num_iterations = max_iter
        self.population = common_initial_population.copy()
        self.best_solution = None
        self.best_fitness = float('inf')
        self.best_fitnesses = []

    # def initialize_population(self):
    #     for i in range(self.pop_size):
    #         particle_position = []
    #         for j in range(self.tasks):
    #             if connections[j]['fns']:  # Check if fog nodes are available for this task
    #                 fn_idx = random.randint(0, len(connections[j]['fns']) - 1)
    #                 particle_position.append(fn_idx)
    #             else:
    #                 particle_position.append(-1)
    #         self.population.append(particle_position)

    def find_worst_solution(self):
        worst_fitness = -1
        worst_idx = -1
        for idx, solution in enumerate(self.population):
            fitness = self.cost_func(solution)
            if fitness > worst_fitness:
                worst_fitness = fitness
                worst_idx = idx
        return worst_idx

    def modify_solution(self, idx):
        current_solution = self.population[idx]
        modified_solution = current_solution.copy()
        for j in range(self.tasks):
            if connections[j]['fns']:  # Check if fog nodes are available for this task
                fn_idx = random.randint(0, len(connections[j]['fns']) - 1)
                modified_solution[j] = fn_idx
        return modified_solution

    def evolve(self):
        for _ in range(self.num_iterations):
            best_fitness = min(self.cost_func(solution) for solution in self.population)
            if best_fitness < self.best_fitness:
                self.best_fitness = best_fitness
                self.best_solution = min(self.population, key=self.cost_func)
            self.best_fitnesses.append(self.best_fitness)
            
            worst_idx = self.find_worst_solution()
            modified_solution = self.modify_solution(worst_idx)
            modified_fitness = self.cost_func(modified_solution)
            if modified_fitness < self.cost_func(self.population[worst_idx]):
                self.population[worst_idx] = modified_solution
            # best_fitness = min(self.cost_func(solution) for solution in self.population)
            # if best_fitness < self.best_fitness:
            #     self.best_fitness = best_fitness
            #     self.best_solution = min(self.population, key=self.cost_func)
            # self.best_fitnesses.append(self.best_fitness)

    def start(self):
        # self.initialize_population()
        self.evolve()

        plt.figure(figsize=(8, 5))
        plt.plot(self.best_fitnesses, color='darkviolet')
        plt.title('Best Fitness over Iterations')
        plt.xlabel('Iteration')
        plt.ylabel('Best Fitness')
        plt.show()

        # print("Optimal assignment:", self.best_solution)
        # print("Minimum cost:", self.best_fitness)

jaya = JayaAlgorithm(pop_size=50, tasks=num_tasks, max_iter=70)
jaya.cost_func = fitness_func
jaya.start()
```

```python
class GSA:
    def __init__(self, pop_size=1, tasks=1, max_iter=100):
        self.cost_func = None

        self.alpha = 0.5
        self.G = 0.5
        self.max_iter = max_iter
        self.pop_size = pop_size
        self.tasks = tasks
        self.best_so_far = None
        self.best_fitnesses=[]

        self.X = np.array(common_initial_population, dtype=float)
        self.V = np.random.rand(pop_size, tasks)
        self.f = np.full((pop_size, tasks), None)  # None  # Will become a list inside cal_f
        self.a = np.full((pop_size, tasks), None)
        self.q = np.full((pop_size, 1), None)
        self.M = np.full((pop_size, 1), None)
        self.cost_matrix = np.full((pop_size, 1), None)
        self.max_delay = None
        self.avg_delay = None

    # # generates initial population
    # def generate_initial_pop(self):
    #     for i in range(self.pop_size):
    #         for j in range(0, self.tasks):
    #             self.X[i][j] = random.uniform(0, len(connections[j]['fns'])-1)

    # Evaluate a single x (x_i)
    def evaluate(self, args):
        return self.cost_func(args)

    # Generate the cost of all particles
    def gen_cost_matrix(self):
        for i, x in enumerate(self.X):
            self.cost_matrix[i] = self.evaluate(x)

    def cal_q(self):
        best = np.min(self.cost_matrix)
        worst = np.max(self.cost_matrix)

        self.q = (self.cost_matrix - worst) / best - worst

    def cal_m(self):
        self.M = self.q / np.sum(self.q)

    def cal_f(self):
        costs = self.cost_matrix.copy()
        costs.sort(axis=0)

        for i in range(self.pop_size):
            f = None
            for cost in costs:
                tmp = np.where(self.cost_matrix == cost)[0]
                j = tmp[0]

                dividend = float(self.M[i][0] * self.M[j][0])
                divisor = np.sqrt(np.sum((self.X[i] - self.X[j]) ** 2)) + np.finfo('float').eps
                if f is None:
                    f = self.G * (dividend / divisor) * (self.X[j] - self.X[i])
                else:
                    f = f + self.G * (dividend / divisor) * (self.X[j] - self.X[i])

            self.f[i] = np.random.uniform(0, 1) * f

    def cal_a(self):
        for i in range(self.pop_size):
            self.a[i] = self.f[i] / self.M[i]

    def cal_v(self):
        self.V = (np.random.uniform(0, 1) * self.V) + self.a

    def move(self):
        self.X = self.X + self.V
        # Round off positions to handle fractional values
        #self.X = np.array([[round(val) for val in row] for row in self.X])
        # Check if positions are within range and adjust if necessary
        for i in range(self.pop_size):
            for j in range(self.tasks):
                if self.X[i][j] < 0:
                    self.X[i][j] = 0
                elif self.X[i][j] >= len(connections[j]['fns']):
                    self.X[i][j] = len(connections[j]['fns']) - 1

    def update_g(self, iteration):
        self.G = self.G * np.e ** (- self.alpha * (iteration / self.max_iter))

    def update_best_so_far(self):
        for x in self.X:
          if self.best_so_far is None or self.evaluate(self.best_so_far) > self.evaluate(x):
              self.best_so_far=x
              
    def start(self):
        _iter = 0
        # self.generate_initial_pop()
        avg = []
        bsf = []
        while _iter < self.max_iter:
            self.gen_cost_matrix()
            self.update_best_so_far()
            avg.append(sum(self.cost_matrix) / len(self.cost_matrix))
            bsf.append(self.evaluate(self.best_so_far))
            
            self.gen_cost_matrix()
            self.cal_q()
            self.cal_m()
            self.update_g(_iter)
            self.cal_f()
            self.cal_a()
            self.cal_v()
            self.move()
            self.update_best_so_far()

            # avg.append(sum(self.cost_matrix) / len(self.cost_matrix))
            # bsf.append(self.evaluate(self.best_so_far))

            _iter += 1
        else:
            plt.figure(figsize=(8, 5))  # Create a figure with a specific size
            plt.plot(bsf, color='darkviolet')
            plt.title('Best So Far Solution')
            plt.xlabel('Iteration')
            plt.ylabel('Best So Far Solution')
            plt.show()
            # print("Optimal assignment:", self.best_so_far)
            # print("Minimum cost:", self.evaluate(self.best_so_far))
        self.best_fitnesses=bsf

gsa = GSA(pop_size=50, tasks=num_tasks, max_iter=70)
gsa.cost_func = fitness_func
gsa.start()
```

```python
class PSO:
    def __init__(self, pop_size=1, tasks=1, max_iter=100):
        self.cost_func = None

        self.w = 0.5  # Inertia weight
        self.c1 = 1.5  # Cognitive weight
        self.c2 = 1.5  # Social weight
        self.max_iter = max_iter
        self.pop_size = pop_size
        self.tasks = tasks
        
        self.best_positions = common_initial_population
        self.velocities = []
        self.personal_best = []
        self.global_best = None
        self.global_best_cost = float('inf')
        
        self.costs = []
        
    def evaluate(self, args):
        return self.cost_func(args)  
        
    # def generate_initial_pop(self):
    #     for i in range(self.pop_size):
    #         particle_position = []
    #         for j in range(self.tasks):
    #             if connections[j]['fns']:  # Check if fog nodes are available for this task
    #                 fn_idx = random.randint(0, len(connections[j]['fns']) - 1)
    #                 particle_position.append(fn_idx)
    #             else:
    #                 particle_position.append(-1)
    #         self.best_positions.append(particle_position)
    #         self.velocities.append(random.randint(-len(connections[j]['fns'])+1, len(connections[j]['fns'])-1))
    #         self.personal_best.append(particle_position)
    #         if self.evaluate(particle_position)<self.global_best_cost :
    #                 self.global_best = particle_position
    #                 self.global_best_cost = self.evaluate(particle_position)


    def update_velocity(self):
        for i in range(self.pop_size):
            r1, r2 = random.random(), random.random()

            # Convert personal_best and best_positions to NumPy arrays
            personal_best_np = np.array(self.personal_best[i])
            best_positions_np = np.array(self.best_positions[i])
    
            cognitive = self.c1 * r1 * (personal_best_np - best_positions_np)
            
            # Convert global_best and best_positions to NumPy arrays
            global_best_np = np.array(self.global_best)
            social = self.c2 * r2 * (global_best_np - best_positions_np)
    
            self.velocities[i] = self.w * self.velocities[i] + cognitive + social
            
    def update_position(self):
        # Convert best_positions and velocities to NumPy arrays
        best_positions = np.array(self.best_positions)
        velocities = np.array(self.velocities)
        self.best_positions=best_positions + self.velocities
        
        # Round off positions to handle fractional values
        #self.best_positions = np.round(self.best_positions)
        
        # Check if positions are within range and adjust if necessary
        for i in range(self.pop_size):
            for j in range(self.tasks):
                if self.best_positions[i][j] < 0:
                    self.best_positions[i][j] = 0
                elif self.best_positions[i][j] >= len(connections[j]['fns']):
                    self.best_positions[i][j] = len(connections[j]['fns']) - 1

    def update_personal_best(self):
        for i in range(self.pop_size):
            cost = self.evaluate(self.best_positions[i])
            if cost < self.evaluate(self.personal_best[i]):
                self.personal_best[i] = self.best_positions[i]
                if cost < self.global_best_cost:
                    self.global_best = self.best_positions[i].copy()
                    self.global_best_cost = cost

    def start(self):
        # self.generate_initial_pop()
        _iter = 0
        while _iter < self.max_iter:
            self.costs.append(self.global_best_cost)
            self.update_velocity()
            self.update_position()
            self.update_personal_best()
            # self.costs.append(self.global_best_cost)  # Append gbest cost, not self.evaluate(self.global_best)
            _iter += 1
        plt.figure(figsize=(8, 5))
        plt.plot(self.costs)
        plt.xlabel('Iteration')
        plt.ylabel('Cost')
        plt.title('PSO Optimization')
        plt.show()
        print("Optimal assignment:", self.global_best)
        print("Minimum cost:", self.global_best_cost)

pso = PSO(pop_size=30, tasks=num_tasks, max_iter=50)
pso.cost_func = fitness_func
pso.start()
```

```python
import numpy as np
import random
from matplotlib import pyplot as plt

class HPSOGA:
    def __init__(self, pop_size=1, tasks=1, max_iter=100, num_fog_nodes=4, num_tasks=10, task_size=[100, 80, 30, 50, 10, 40, 20, 30, 50, 15], queue_length=5):
        self.cost_func = None
        self.pop_size = pop_size
        self.tasks = tasks
        self.max_iter = max_iter
        self.num_fog_nodes = num_fog_nodes
        self.num_tasks = num_tasks
        self.task_size = task_size
        self.queue_length = queue_length
        self.mu_f = np.round(np.random.uniform(1, 10, num_fog_nodes), 2)  # fixed traffic rate for each fog node
        self.varphi_f = 0.1 * self.mu_f  # 10% incoming rate is transferred
        self.lambda_f = np.zeros(num_fog_nodes)
        for i in range(num_fog_nodes):
            self.lambda_f[i] = np.round(np.random.uniform(1, self.mu_f[i]), 2)

        # PSO parameters
        self.alpha = 0.5
        self.G = 0.1
        self.best_so_far = None

        self.X = np.random.rand(pop_size, tasks)
        self.V = np.random.rand(pop_size, tasks)
        self.f = np.full((pop_size, tasks), None)
        self.a = np.full((pop_size, tasks), None)
        self.q = np.full((pop_size, 1), None)
        self.M = np.full((pop_size, 1), None)
        self.cost_matrix = np.full((pop_size, 1), None)

        # GA parameters
        self.p_e = 0.2  # Probability of applying mutation
        self.p_c = 0.8  # Probability of applying crossover
        self.t_max = max_iter  # Maximum number of iterations

    def generate_initial_pop(self):
        for i in range(self.pop_size):
            for j in range(self.tasks):
                self.X[i][j] = random.uniform(0, self.num_fog_nodes - 1)

    def evaluate(self, args):
        return self.cost_func(args)

    def gen_cost_matrix(self):
        for i, x in enumerate(self.X):
            self.cost_matrix[i] = self.evaluate(x)

    def cal_q(self):
        best = np.min(self.cost_matrix)
        worst = np.max(self.cost_matrix)
        self.q = (self.cost_matrix - worst) / best - worst

    def cal_m(self):
        self.M = self.q / np.sum(self.q)

    def cal_f(self):
        costs = self.cost_matrix.copy()
        costs.sort(axis=0)
        for i in range(self.pop_size):
            f = None
            for cost in costs:
                tmp = np.where(self.cost_matrix == cost)[0]
                j = tmp[0]
                dividend = float(self.M[i][0] * self.M[j][0])
                divisor = np.sqrt(np.sum((self.X[i] - self.X[j]) ** 2)) + np.finfo('float').eps
                if f is None:
                    f = self.G * (dividend / divisor) * (self.X[j] - self.X[i])
                else:
                    f = f + self.G * (dividend / divisor) * (self.X[j] - self.X[i])
            self.f[i] = np.random.uniform(0, 1) * f

    def cal_a(self):
        for i in range(self.pop_size):
            self.a[i] = self.f[i] / self.M[i]

    def cal_v(self):
        self.V = (np.random.uniform(0, 1) * self.V) + self.a

    def move(self):
        self.X = self.X + self.V
        for i in range(self.pop_size):
            for j in range(self.tasks):
                if self.X[i][j] < 0:
                    self.X[i][j] = 0
                elif self.X[i][j] >= self.num_fog_nodes:
                    self.X[i][j] = self.num_fog_nodes - 1

    def update_g(self, iteration):
        self.G = self.G * np.e ** (-self.alpha * (iteration / self.max_iter))

    def update_best_so_far(self):
        for x in self.X:
            if self.best_so_far is None or self.evaluate(self.best_so_far) > self.evaluate(x):
                self.best_so_far = x

    def roulette_wheel_selection(self, population, fitness_values):
        probabilities = fitness_values / np.sum(fitness_values)
        selected_index = np.random.choice(len(population), p=probabilities)
        return population[selected_index]

    def mutation(self, solution):
        index = np.random.randint(len(solution))
        new_value = np.random.randint(self.num_fog_nodes)
        solution[index] = new_value
        return solution

    def crossover(self, parent1, parent2):
        crossover_point = np.random.randint(1, self.tasks)
        child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
        child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
        return child1, child2

    def evolve(self):
        CH_t = []
        for m in range(self.pop_size):
            if np.random.rand() < self.p_e:
                parent = self.roulette_wheel_selection(self.X, self.cost_matrix)
                offspring = self.mutation(parent)
                CH_t.append(offspring)

            if np.random.rand() < self.p_c:
                parent1 = self.roulette_wheel_selection(self.X, self.cost_matrix)
                parent2 = self.roulette_wheel_selection(self.X, self.cost_matrix)
                child1, child2 = self.crossover(parent1, parent2)
                CH_t.append(child1)
                CH_t.append(child2)

        combined_population = []
        for child in CH_t:
            child_fitness = self.evaluate(child)
            for i, (solution, fitness) in enumerate(zip(self.X, self.cost_matrix)):
                if child_fitness < fitness:
                    combined_population.append(child)
                    break
                elif np.array_equal(solution, self.best_so_far):
                    combined_population.append(solution)
                    break

        self.X = combined_population

    def start(self):
        self.generate_initial_pop()
        avg_fitness = []
        best_fitness = []
        for iter in range(self.t_max):
            self.gen_cost_matrix()
            self.cal_q()
            self.cal_m()
            self.update_g(iter)
            self.cal_f()
            self.cal_a()
            self.cal_v()
            self.move()
            self.update_best_so_far()

            avg_fitness.append(sum(self.cost_matrix) / len(self.cost_matrix))
            best_fitness.append(self.evaluate(self.best_so_far))

        plt.figure(figsize=(8, 5))
        plt.plot(best_fitness, color='darkviolet')
        plt.title('Best So Far Solution')
        plt.xlabel('Iteration')
        plt.ylabel('Best So Far Solution')
        plt.show()

        print("Optimal assignment:", self.best_so_far)
        print("Minimum cost:", self.evaluate(self.best_so_far))

hpsoga = HPSOGA(pop_size=50, tasks=num_tasks, max_iter=70)
hpsoga.cost_func = fitness_func
hpsoga.start()
```
